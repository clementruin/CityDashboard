```python
import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize

import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
```

```python
# open 1st dataframe
with open('static_dic/budgets_idf.json') as file:
    data = json.load(file)

df1 = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df1.set_index('code', inplace=True)
```

```python
# open second dataframe
with open('static_dic/chomage.json') as file:
    data = json.load(file)

df2 = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df2.set_index('code', inplace=True)
```

```python
# open third dataframe
with open('static_dic/partisidf.json') as file:
    data = json.load(file)

df3 = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df3.set_index('code', inplace=True)
```

```python
# open fourth dataframe
with open('static/logement.json') as file:
    data = json.load(file)

df4 = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df4.set_index('inseecode', inplace=True)
```

```python
# concatenate dataframes 
df = pd.concat([df1, df2, df3, df4], axis=1, join='inner')
```

```python
# We drop NAs for the analysis of parties
df = df[df['partis.2001'] != 'NA']
df = df[df["partis.2008"] != 'NA']
df = df[df["partis.2014"] != 'NA']

# Encode categorical features
encoder = {"partis.2001" : {"FN" : 8,
                            "UMP-LR" : 5,
                            "DVD" : 3,
                            "UDI" : 1,
                            "SE" : 0,
                            "MoDem" : 0,
                            "PRG" : -1,
                            "EELV" : -1,
                            "DVG" : -3,
                            "PS" : -5,
                            "FG" : -7,
                            "PCF" : -8},
            "partis.2008" : {"FN" : 8,
                            "UMP-LR" : 5,
                            "DVD" : 3,
                            "UDI" : 1,
                            "SE" : 0,
                            "MoDem" : 0,
                            "PRG" : -1,
                            "EELV" : -1,
                            "DVG" : -3,
                            "PS" : -5,
                            "FG" : -7,
                            "PCF" : -8},
           "partis.2014" : {"FN" : 8,
                            "UMP-LR" : 5,
                            "DVD" : 3,
                            "UDI" : 1,
                            "SE" : 0,
                            "MoDem" : 0,
                            "PRG" : -1,
                            "EELV" : -1,
                            "DVG" : -3,
                            "PS" : -5,
                            "FG" : -7,
                            "PCF" : -8}
          }

df.replace(encoder, inplace=True)
df.head()
```

```python
# Select the most imprtant cities
df = df[df['housing'] > 2000]
```

```python
# Features
features = list(df.columns)[-10:]
print(features)
```

```python
correlations = df[features].corr(method='pearson')

fig = plt.figure(figsize=(30,30))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, cmap='RdBu',vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(features),1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(features, rotation='vertical')
ax.set_yticklabels(features)
plt.show()
```
