```python
import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
from sklearn import cluster
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
```

```python
# open Budget dataframe
with open('static_dic/budgets_idf.json') as file:
    data = json.load(file)

df_budget = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_budget.set_index('code', inplace=True)
```

```python
# open Chomage dataframe
with open('static_dic/chomage.json') as file:
    data = json.load(file)

df_chomage = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_chomage.set_index('code', inplace=True)
```

```python
# open Partis dataframe
with open('static_dic/partisidf.json') as file:
    data = json.load(file)

df_partis = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_partis.set_index('code', inplace=True)
```

```python
# open Logmement dataframe
with open('static/logement.json') as file:
    data = json.load(file)

df_logement = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_logement.set_index('inseecode', inplace=True)
```

```python
# open Bac dataframe
with open('static_dic/Bac_js2.json') as file:
    data = json.load(file)

df_bac = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_bac.set_index('Code', inplace=True)
```

```python
# open Baccalaureat dataframe
with open('static/logement.json') as file:
    data = json.load(file)

df_bac = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_bac.set_index('inseecode', inplace=True)
```

```python
# open Siren dataframe
with open('static_dic/sirene_93_01_01_18.json') as file:
    data = json.load(file)

df_siren = pd.DataFrame.from_dict(json_normalize(data), orient='columns')
df_siren.set_index('Insee', inplace=True)

df_siren.head(40)
```

```python
# concatenate dataframes 
df = pd.concat([df_bac, df_budget, df_chomage, df_logement, df_partis], axis=1, join='outer')
```

```python
# We drop NAs for the analysis of parties
print('Before drop NA partis :',  df.shape)
df = df[df['partis.2001'] != 'NA']
df = df[df["partis.2008"] != 'NA']
df = df[df["partis.2014"] != 'NA']
print('After drop NA partis :',  df.shape)

# Encode categorical features
encoder = {"partis.2001" : {"FN" : 8,
                            "UMP-LR" : 5,
                            "DVD" : 3,
                            "UDI" : 1,
                            "SE" : 0,
                            "MoDem" : 0,
                            "PRG" : -1,
                            "EELV" : -1,
                            "DVG" : -3,
                            "PS" : -5,
                            "FG" : -7,
                            "PCF" : -8},
            "partis.2008" : {"FN" : 8,
                            "UMP-LR" : 5,
                            "DVD" : 3,
                            "UDI" : 1,
                            "SE" : 0,
                            "MoDem" : 0,
                            "PRG" : -1,
                            "EELV" : -1,
                            "DVG" : -3,
                            "PS" : -5,
                            "FG" : -7,
                            "PCF" : -8},
           "partis.2014" : {"FN" : 8,
                            "UMP-LR" : 5,
                            "DVD" : 3,
                            "UDI" : 1,
                            "SE" : 0,
                            "MoDem" : 0,
                            "PRG" : -1,
                            "EELV" : -1,
                            "DVG" : -3,
                            "PS" : -5,
                            "FG" : -7,
                            "PCF" : -8}
          }

df.replace(encoder, inplace=True)
df.head()
```

```python
# Features
#features = list(df.columns)
features = ['partis.2014', 
            'chomage', 
            'housing',
            'main_res', 
            'portion_hlm_tenant',
            'budget_2015', 
            'd1.2_2015', 
            'd1.3_2015', 
            'd1.4_2015', 
            'd1.5_2015', 
            'd1_2015',
            'd2.1_2015',
            'd2.2_2015',
            'd2.3_2015',
            'd2_2015',
            'depenses_2015',
            'r1.1_2015',
            'r1.2_2015',
            'r1.3_2015',
            'r1_2015',
            'r2.1_2015',
            'r2.2_2015',
            'r2_2015',
            'recettes_2015',
            'NbLycees_2017',
            'Priv_2017',
            'Public_2017',
            'Resultat_2017_(50)',
            'Resultat_2017_(50-70)',
            'Resultat_2017_(70-80)',
            'Resultat_2017_(80-90)',
            'Resultat_2017_(90)'
           ]
print(features)

df = df[features]
```

```python
# Select the most imprtant cities
df = df[df['housing'] > 5000]

# get rid of NaNs
df = df.dropna(axis=0, how='any')
```

```python
correlations = df[features].corr(method='pearson')

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, cmap='RdBu',vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(features),1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(features, rotation='vertical')
ax.set_yticklabels(features)
plt.show()
```

```python
# Clean NaN's before clustering 

df.dropna(axis=0, how='any')
df.info()
df.describe()

test=df.as_matrix(columns=df.columns)
np.isnan(test).any()
```

```python
# KMeans clusters

df = df[features]
kmeans = cluster.KMeans(n_clusters=2, random_state=0).fit(df)
print(kmeans.labels_)

# Normalize data

df = df.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
```

```python
# Plot 2D 
plt.figure()
plt.subplot(221)
plt.scatter(df["partis.2014"], df["chomage"], c=kmeans.labels_)
plt.subplot(222)
plt.scatter(df["partis.2014"], df['budget_2015'], c=kmeans.labels_)
plt.subplot(223)
plt.scatter(df["chomage"], df['budget_2015'], c=kmeans.labels_)

# Plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["partis.2014"], df["chomage"], df['budget_2015'], c=kmeans.labels_)
```
